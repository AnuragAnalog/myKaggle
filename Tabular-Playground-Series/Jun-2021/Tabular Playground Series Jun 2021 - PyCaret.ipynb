```python
# Required modules

import numpy as np
import pandas as pd

from zipfile import ZipFile
from pycaret.classification import *

from sklearn.metrics import mean_squared_log_error
```

```python
# Extracting the data

with ZipFile('./tabular-playground-series-jun-2021.zip', 'r') as zf:
    zf.extractall('./')
```

```python
# Load the train data

train = pd.read_csv('train.csv')
train.head()
```

```python
# Inspecting the data

train.info()
train.describe()
```

```python
# Load test data

test = pd.read_csv('test.csv')
test.head()
```

```python
# Missing values

if train.isna().any().any():
    print(train.isna().sum()/train.shape[0])
else:
    print("No Missing values")
```

```python
# Setting Up and adding metric

setup(data=train,
    target='target',
    normalize = True,
    normalize_method = 'robust',
    silent= True,
    use_gpu=False,
    n_jobs=-1,
    session_id = 42,
    numeric_imputation = 'mean')

try:
    add_metric('msle', 'MSLE', mean_squared_log_error, greater_is_better=False, target='pred_proba')
except:
    pass
```

```python
# Creating Models

fold = 5

model_names = ['lightgbm', 'catboost']
model_list = list()

for model_name in model_names:
    model = create_model(estimator=model_name, fold = fold)
    model_list.append(model)

# Blending the model

blended = blend_models(estimator_list=model_list, fold=fold)
pred_holdout = predict_model(blended)

final_model = finalize_model(blended)
```

```python
# Finalize the model

prep_pipe = get_config('prep_pipe')
prep_pipe.steps.append(['trained_model', final_model])
test_pred = prep_pipe.predict_proba(test)
test_pred
```

```python
# Output

submission = pd.DataFrame(test_pred)
submission.columns = sorted(pd.unique(train['target']))
submission['id'] = test['id'].values
submission.to_csv('output.csv', index=False)
```
